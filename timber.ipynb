# Supporting a wall and point load

## Values that have been given for this study

```python
Length = 8.5; Width = 7; Gap = 5.5; #All measurements in meters
```

###Load on the wall due to bricks

```python
Wb = Gap * 3 * 0.12 * 1922 * 9.81 #Wb is the weight due to the brick wall
```

```python
Wb
```

###Ww is the dead load from wall on G beam. Which in this instance is the Uniform distributed load

##### The factor of safety here is 1.35 for a variable load we could have used 1.5

```python
Ww = Wb *1.35
```

```python
Ww
```

##The maximum moment generated by the wall is Mmax 

```python
Mmax = Ww * Gap/8 + 40000*G/4 #These are standard formulae for a UDL and a point load
Mmax
```

##The section modulus calculation table 11.18 $W_{el} > \frac{M_{max}}{\sigma}$

### Consider the use of C30 timber which has $\sigma = 11$ (bending stress parallel to plane $N/mm^2$)

```python
Wel = Mmax/11
Wel
```

###Checking to see if deflection is smaller than permissable amount $\delta = \Sigma c \frac{W \times L^3}{E\times I}$

```python
E = 8200;I = 4500; 
#E is elasticity from table 11.11, I is the second moment of inertia from table 11.18 highest available used for 300x300
```

```python
delta = 5/384*Ww*Gap**3/(E*I)+ 1/48*40000*Gap**3/(E*I)
delta
```

###Maximum $\delta_{permissable}$ = 0.003*Gap

```python
0.003*Gap
```
