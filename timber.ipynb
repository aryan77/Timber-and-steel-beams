#Timber beam 
###Values that have been given for this study
All measurements in metres

```python
Length = 8.5;  Width = 7;   Gap = 5.5;
```

###Load on the beam due to the bricks in wall
$W_b$ is the weight due to the brick wall measured in Newtons. 9.81 is the acceleration due to gravity.
$W_b$ is calculated by working out the volume of the brick wall, multiplied by density, multiplied by gravity.

```python
Wb = Gap * 3 * 0.12 * 1922 * 9.81 
Wb
```

###$W_w$ is the dead load from wall on G beam.
The factor of safety here for a dead load is 1.35 for a variable live load we would use 1.5

```python
Ww = Wb *1.35
Ww
```

###The maximum moment generated by the wall is $M_{max}$ 
Standard formulae for a UDL and a point load have been used to work out the combined moment generated by the timber and the point load itself.

```python
Mmax = Ww * Gap/8 + 40000*Gap/4 #
Mmax
```

### Calculation of the section modulus $W_{el}$ for a timber support  
$$W_{el} > \frac{M_{max}}{\sigma}$$ 
Consider the use of C30 timber which has $\sigma = 11$ ($\sigma$ is the bending stress parallel to plane $N/mm^2$, ref table 11.18 for values)

```python
Wel = Mmax/11
Wel
```

Unfortunately there is no timber with a $W_{el}$  $x-x$ section modulus bigger than 8150mm to accomodate this magnitude of stress applied to the C30 timber. However for completeness I shall continue to work out the deflection $\delta$ and check to see if it is less than  $\delta_{permissable}$ = 0.003*Gap

###Timber Deflection calculation
$$\delta = \Sigma C \frac{W \times L^3}{E\times I}$$
$E$ is the minumum modulus of elasticity from table 11.11 of C30 timber, $I$  $x-x$ is the second moment of inertia from table 11.18 highest available used for 300mmx300mm

```python
E = 8200;  I = 675; 
```

The standard loading coefficients for each fraction term, $C$, have been taken from lectures in Week 21. For a uniformly distributed load $C = \frac{5}{384}$ for a point load $C = \frac{1}{48}$

```python
delta = 5/384*Ww*Gap**3/(E*I)+ 1/48*40000*Gap**3/(E*I)
delta
```

###Maximum permissible deflection of timber beam 
$$\delta_{permissible} = 0.003\times\text{Gap}$$

```python
delta_permissible = 0.003*Gap
delta_permissible
```

The deflection $\delta =0.04477$m is greater than the permissable amount $\delta_{permissible} = 0.0165$m; in addition, there is no timber with a section modulus, $W_{el}$ larger than 8150mm. Therefore I shall now consider the use of a steel beam.

##Timber joists analysis

The total load carried by each joist is $W_j$ where 
$$W_j = \frac{L}{2}\times \text{spacing (m)} \times \text{load(N)} \times \text{safety factor}$$

```python
Wj = 0.5*Length*0.4*1.2*10**3*1.35
Wj
```

Since the moment capacity is $M_{max} = \sigma W_{el}$ we have have that $$\sigma W_{el} = \frac{L}{2}\times W_j \times\frac{1}{8}$$ rearranging this formula gives $$W_{el} = \frac{L\times W_j}{16\sigma}$$

Let us examine the possibility of using C24 timber with $\sigma= 7.5$

```python
joist_Wel = Length * Wj/(16*7.5)
joist_Wel
```

By looking at table 11.18 it can be seen that C24 timber size 75x150 has a section modulus of 281 (>195.075) which means this choise is safe for use. 
However we must now check for deflection caused by the use of C24 75x150.
###Timber joist deflection calculation for C24 75x150
$$\delta = \Sigma C\frac{W \times Span^3}{E\times I}$$
The load is uniformly distributed over the joist so $C = \frac{5}{384}$, thus $$\delta = \frac{5}{384}\times\frac{W_j \times (\frac{L}{2})^3}{E\times I}$$
Maximum permissible delta is $$\delta_{permissible} = 0.003\times\frac{L}{2}$$

```python
delta_permissible = 0.003*Length/2
delta_permissible
```

For joist C24 $E = 7200$ see table 11.11, for size 75x150 $I_{x-x} = 21.1$ see table 11.18

```python
delta_joistC24 = 5/384*Wj*(Length/2)**3/(7200*21.1)
delta_joistC24
```

The above calculation shows us that timber C24 of size 75x150 is not safe for use because there is too much deflection $$\delta > \delta_{permissible}$$ i.e. 0.01812 > 0.01275 

Here we have two options, to find a suitable timber joist. 

We can choose a higher grade timber e.g. C30 and then repeat the analysis using the correct values for $E$ and $\sigma$ i.e. $E = 8200$ and $\sigma=11.0$ . Thereafter we would need to choose a size of beam that has a section modulus greater than the newly calculated $W_{el}$ and repeat the deflection check.

It is simpler however to resize our C24 beam (thus keeping $E = 7200$) and finding a higher inertia value of $I$.
For a C24 timber of size 75x175 $I=33.5$.

```python
delta_joistC24 = 5/384*Wj*(Length/2)**3/(7200*33.5)
delta_joistC24
```

This shows us that the deflection is now less than  $\delta_{permissible}$ i.e. 0.01141 < 0.01275 and so we can safely use a C24 joist of size 75x175 which has ample section modulus 383 (>195) and a small enough deflection of 0.01141 (< 0.01275).
